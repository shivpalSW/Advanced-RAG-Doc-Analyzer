```python
import logging
import os 
from datetime import datetime
```

```python
# Create a directory for logs if it doesn't exist
logs_dir= os.path.join(os.getcwd(), "logs")
```

```python
os.makedirs(logs_dir, exist_ok=True)
```

```python
# today
# with today's date log file will be created
#LOG_FILE= f"{logs_dir}/log_{datetime.now().strftime('%Y-%m-%d_%H-%M-%S')}.log"
#this is a log file(name of the log file would current data and time)
LOG_FILE = f"{datetime.now().strftime('%d-%m-%Y_%H-%M-%S')}.log"
LOG_FILE
```

```python

#today
# with today's date log file will be creted

#tomorrow
# with tomorrow's date log file will be created
```

```python
LOG_FILE_PATH=os.path.join(logs_dir, LOG_FILE)
```

```python
LOG_FILE_PATH
```

```python
# Configure logging 
logging.basicConfig(
    filename=LOG_FILE_PATH,
    format="[ %(asctime)s ] %(levelname)s %(name)s (line:%(lineno)d) - %(message)s",
    level=logging.INFO,
)
```

```python
logger= logging.getLogger("Document portal")
```
